```python
#Sebastian Acevedo 22258530
from random import choice, randint
import pandas as pd

# Productos y estudiantes de ejemplo

#materias = ["fisica mecanica", "estructura de datos", "programacion con php", "cts", "Calculo integral"]
estudiantes = ["hernesto", "Juan", "Sebastian", "ivan", "pedro","sandra", "andrea"]
apellidos = ["Acevedo","Restrepo","Gaviria", "Vargas", "Munoz"]


# Generar 100 estudiante aleatorias
estudiante = []
for _ in range(5):
    nomEstudiante = choice(estudiantes)
    apellidoEst = choice(apellidos)
    nota = randint(0,5)
    mate = {}
    #for i in materias:
     #   mate[i]= randint(0,5)
    est = {
        "apellido": apellidoEst,
        "nomEstudiante": nomEstudiante,
        "fisica mecanica": randint(0,5),
        "estructura de datos": randint(0,5),
        "programacion con php": randint(0,5),
        "cts": randint(0,5),
        "Calculo integral": randint(0,5)

    }


    estudiante.append(est)
estudianteDF = pd.DataFrame(estudiante)
pd.DataFrame(estudiante)
```
